### Performing preliminary data analysis on RAW game data.

```python
import pandas as pd
import numpy as np

# Your data names (suffixes) here:
match_data_file_name = ''
player_data_file_name = ''
load_path = '../data/raw/'
save_path = '../data/processed/'

match_data = pd.read_csv(load_path + 'match_data_' + match_data_file_name + '.csv')
player_data = pd.read_csv(load_path + 'players_' + player_data_file_name + '.csv')

match_data
```

Finding all players present in all games in game_data:

```python
combined_player_names = match_data.loc[:, ['p1_name', 'p1_puuid', 'p1_summonerId']]
combined_player_names = combined_player_names.rename(columns = {'p1_name': 'name',
                                                                'p1_puuid': 'puuid',
                                                                'p1_summonerId': 'summonerId'})

for i in range(2,11):
    
    player_names = match_data.loc[:, ['p{j}_name'.format(j = i),
                                        'p{j}_puuid'.format(j = i), 
                                        'p{j}_summonerId'.format(j = i)]]
                                        
    player_names = player_names.rename(columns = {'p{j}_name'.format(j = i): 'name',
                                                  'p{j}_puuid'.format(j = i): 'puuid',
                                                  'p{j}_summonerId'.format(j = i): 'summonerId'})

    combined_player_names = pd.concat([combined_player_names, player_names], ignore_index = True)
    combined_player_names = combined_player_names.drop_duplicates( ignore_index = True)
    

combined_player_names
```

Saving combined_player_names to .csv file in ../data/processed/: 

```python
combined_player_names.to_csv(save_path + 'players_{name}.csv'.format(name = match_data_file_name), index = False)
```
